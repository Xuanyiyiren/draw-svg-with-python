```python
from svgwrite import *
```

# 创建一条直线

```python
dwg = Drawing('fig/01_lines.svg',profile='tiny',size=('1200','400'),viewBox="0 0 1200 400")
dwg.add(dwg.rect(insert=(1,1),size=(1200-2,400-2),fill='none',stroke='blue',stroke_width=2))
g = dwg.g(stroke='orange')

w0=5
a0=100
for _ in range(5):
    g.add(dwg.line(start=(a0,300),end=(a0+200,100),stroke_width=w0))
    # print(w0)
    a0+=200
    w0+=5
dwg.add(g)
dwg.save()
dwg
```

第一行代码创建了Drawing对象dwg
- 其filename是`'fig/01_lines_1.svg'`，当使用save()属性保存时，就会以这个名字生成svg图片。
- profile[是svg图片的一种配置](https://developer.mozilla.org/en-US/docs/Web/SVG/Attribute/baseProfile)
- size定义了svg图片的大小
- viewBox定义了svg图片的视窗，**前两个数字代表视窗的起点（左上角），后两个数字代表始创的宽度和高度**

第二行代码添加了一个矩形边框

第三行代码是创建了一个group对象，下面的for循环中，依次添加了不同宽度的线条

接着是将group对象添加到dwg中并保存


```python
dwg1 = Drawing('fig/01_lines_1.svg',profile='tiny',size=('1200','400'),viewBox="0 0 1200 400")
dwg1.add(dwg.rect(insert=(1,1),size=(1200-2,400-2),fill='none',stroke='red',stroke_width=2))

dwg1.add(g)
dwg1.save()
dwg1
```

这说明不同的Drawing对象创建的line,rect等几何对象是一致的。

# 绘制圆

```python
dwg2 = Drawing('fig/02_circle1.svg',profile='tiny',viewBox="-50 -50 100 100")
dwg2.add(dwg2.rect(insert=(-50+1,-50+1),size=(100-2,100-2),fill='none',stroke='blue',stroke_width=2))

circle = dwg2.circle(center=(0,0),r=30,fill='red')
dwg2.add(circle)
circle = dwg2.circle(center=(10,10),r=10,fill='none',stroke='purple',stroke_width=2)
dwg2.add(circle)

dwg2.save()
dwg2
```

# 折线图

```python
from random import randint
dwg4 = Drawing('fig/04_line_graph.svg',profile='tiny',viewBox='0 5 300 205')
points=[(i*30,randint(10,200)) for i in range(10)]
dwg4.add(dwg.polyline(points,stroke='blue',fill='none'))
dwg4.save()
dwg4
```

## 密的折线图等于函数图 

```python
import numpy as np
size=(600,300)
viewBox=str(0-10)+' '+str(-size[1]/2-10)+' '+str(size[0]+20)+' '+str(size[1]+20)
x = np.linspace(0,2*np.pi,1000)
y = np.sin(x)

show_x = x/(2*np.pi)*size[0]
show_y = y/2*size[1]

points_ = list(zip(show_x,show_y))


dwg5 = Drawing('fig/05_sin.svg',profile='tiny',viewBox=viewBox)
dwg5.add(dwg.polyline(points_,stroke='blue',fill='none'))
dwg5.save()
dwg5
```
